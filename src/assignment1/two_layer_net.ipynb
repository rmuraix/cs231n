In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?

## Answer:

**SigmoidとReLU**

シグモイドとReLUはグラフにすると以下のようになる。

シグモイド関数に大きな正または負の入力を与えると勾配は0になってしまう。またシグモイド関数の微分の最大値は0.25になる。このため隠れ層を遡るごとに、(活性化関数の微分がかけ合わさる)伝播していく誤差はどんどん小さくなっていく。この結果、多くの隠れ層がある場合には入力層近くの隠れ層まで誤差をフィードバックできない。（**勾配消失問題**）

ReLU(Rectified Linear Unit, 正規化線形ユニット)関数の場合は$x > 0$のとき微分値は1である。一方で$x \le 0$のとき微分値は0になる。

一方、Leaky ReLUでは$x \le 0$でもわずかに傾斜がついており勾配が0になることはない。

以上から答えはSigmoidとReLUになる。

> これらの関数を1次元のケースで考えた場合、どのような入力がこのような振る舞いにつながるでしょうか？

シグモイド:　大きな正または負の入力を受けたとき勾配が0に非常に近くなる。

ReLU: 負の値を受け取ったとき勾配は0になる。

1. より大きなデータセットでトレーニングする。
2. 隠れユニットを増やす。
3. 正則化の強さを上げる。
4. 上記のどれでもない。

$\color{blue}{\textit Your Answer:}$

`1`と`3`

$\color{blue}{\textit Your Explanation:}$

1: データが多ければ多いほど、ネットワークは特定の入力値の重みを学習する。重みが優れているほど、より正確な特徴がネットワークのパフォーマンスに影響を与える。従って、モデルはより良く汎化され、訓練データにオーバーフィットしなくなる。しかし、類似したサンプルは新しい情報を提供しないため、より大きなデータセットにはできるだけ様々なサンプルを含めるべきである。

2: 隠れユニットを増やしてもオーバーフィッティングを解決することはできない。なぜなら、モデルがより小さなパターンを学習し、学習データに含まれる可能性のあるアーティファクトを学習してしまうからだ。

3: 正則化項を大きくして重みにペナルティを与えることで、特定の重みが特定の特徴に影響を与えすぎないようになり、モデルのロバスト性が向上する。
