### Inline Question 1:

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:

Functions that have this problem: `1`, `2`

Explanation:
* _Sigmoid_ may saturate and kill gradients - if neuron's activation becomes close to `1` or `0`, its gradient becomes `0` which prevents further parameter updates. This might happen if data is unnormalized or absolute weight values become too large.
  * Aside: when large inputs are passed through network layers, gradients may become imbalanced and cause _exploding gradient_.
* _ReLU_ might die and stop allowing updates - if all inputs cause the activation to be `0`, the gradient with respect to either input would become `0`. This could happen due to large learning rate, which could allow some input batch to suddenly drive weights to large negative values.

Examples in 1D:
* _Sigmoid_: if it receives a large positive or negative input (e.g., `1e5`), then according to formula, $\sigma(x)=\frac{1}{1+e^{-x}}$, the output would be very close to `0` or `1`. Furthermore, according to derivative formula, $\nabla_x\sigma=(1-\sigma(x))\sigma(x)$, it is obvious that the gradient woud be very close to `0`.
* _ReLU_: according to formua, $\text{ReLU}(x)=\max(0,x)$, if it receives any negative value (e.g., `-1`), the output will be `0` thus the gradient will also be `0`.

1. シグモイド

## 答え

この問題がある関数： `1`, `2`

説明
* もしニューロンの活性化が `1` や `0` に近づくと、その勾配は `0` となり、それ以上パラメータを更新することができなくなります。これはデータが正規化されていなかったり、重みの絶対値が大きすぎたりすると起こるかもしれません。
  * 余談：大きな入力がネットワーク層に渡されると、勾配のバランスが崩れて _exploding gradient_ が発生することがあります。
* すべての入力が活性化を `0` にする場合、どちらかの入力に対する勾配は `0` になります。これは学習速度が大きいために起こる可能性があり、いくつかの入力バッチが突然重みを大きな負の値に追いやる可能性があります。

1次元での例

* _ReLU_: 式によると、$text{ReLU}(x)=max(0,x)$, 負の値(例えば`-1`)を受け取ると、出力は`0`になり、勾配も`0`になる。

Ways to decrease the gap: `1`, `3`

$\color{blue}{\textit Your Explanation:}$

If the gap between the training data and test data is large, the model is _overfitting_. Only `1` and `3` are possible solutions:
1. **TRUE** - the more data there is, the better the network learns to weigh certain input values. The better the weights are, the more exact features have impact on a network's performance. Thus the model generalizes better and does not overfit the training data. The larger dataset, however, should include as varying samples as possible because similar samples do not provide new information.
2. **FALSE** - adding more hidden units does not solve _overfitting_ because it allows the model to learn more tiny patterns and possible artifacts in the training data.
3. **TRUE** - penalizing the weights with larger regularization terms would improve the model's robustness because that would not allow certain weights to have too much impact on some specific feature.
