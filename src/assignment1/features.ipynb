## Load data
Similar to previous exercises, we will load CIFAR-10 data from disk.

これまでの練習と同様に、CIFAR-10のデータをディスクから読み込みます。

```python
from cs231n.features import color_histogram_hsv, hog_feature


def get_CIFAR10_data(num_training=49000, num_validation=1000, num_test=1000):
    # CIFAR-10の生データを読み込む
    cifar10_dir = "cs231n/datasets/cifar-10-batches-py"

    # データを何度も読み込まないよう、変数をクリーンアップする（メモリの問題を引き起こす可能性がある）
    try:
        del X_train, y_train
        del X_test, y_test
        print("Clear previously loaded data.")
    except:
        pass

    X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

    # データのサブサンプル
    mask = list(range(num_training, num_training + num_validation))
    X_val = X_train[mask]
    y_val = y_train[mask]
    mask = list(range(num_training))
    X_train = X_train[mask]
    y_train = y_train[mask]
    mask = list(range(num_test))
    X_test = X_test[mask]
    y_test = y_test[mask]

    return X_train, y_train, X_val, y_val, X_test, y_test


X_train, y_train, X_val, y_val, X_test, y_test = get_CIFAR10_data()
```

## Extract Features
For each image we will compute a Histogram of Oriented
Gradients (HOG) as well as a color histogram using the hue channel in HSV
color space. We form our final feature vector for each image by concatenating
the HOG and color histogram feature vectors.

Roughly speaking, HOG should capture the texture of the image while ignoring
color information, and the color histogram represents the color of the input
image while ignoring texture. As a result, we expect that using both together
ought to work better than using either alone. Verifying this assumption would
be a good thing to try for your own interest.

The `hog_feature` and `color_histogram_hsv` functions both operate on a single
image and return a feature vector for that image. The extract_features
function takes a set of images and a list of feature functions and evaluates
each feature function on each image, storing the results in a matrix where
each column is the concatenation of all feature vectors for a single image.

各画像について、HOG（Histogram of Oriented Gradients）ヒストグラムと、HSVの色相チャンネルを使ったカラーヒストグラムを計算する。
色空間の色相チャンネルを使用したカラーヒストグラムを計算します。HOGとカラーヒストグラムの特徴ベクトルを連結して、各画像の最終的な特徴ベクトルを作成します。
HOGとカラーヒストグラムの特徴ベクトルを連結します。

大雑把に言えば、HOGは画像のテクスチャをとらえ、色情報は無視します。
色ヒストグラムは入力画像の色を表し、テクスチャは無視する。
カラーヒストグラムはテクスチャを無視して入力画像の色を表す。その結果、両者を併用することで
を併用した方が、どちらかを単独で使うよりもうまくいくはずです。この仮定を検証する
この仮定を検証することは、あなた自身の興味のために試してみる良いことです。

`hog_feature`関数と`color_histogram_hsv`関数は、どちらも1つの画像に対して処理を行い、その画像の特徴ベクトルを返します。
その画像の特徴ベクトルを返す。extract_features
関数は、画像の集合と特徴関数のリストを受け取り、それぞれの画像に対して各特徴関数を評価します。
その結果を行列に格納します。
各列は，1つの画像に対するすべての特徴ベクトルを連結したものです。

```python
from cs231n.features import *

num_color_bins = 10  # カラーヒストグラムのビンの数
feature_fns = [hog_feature, lambda img: color_histogram_hsv(img, nbin=num_color_bins)]
X_train_feats = extract_features(X_train, feature_fns, verbose=True)
X_val_feats = extract_features(X_val, feature_fns)
X_test_feats = extract_features(X_test, feature_fns)

# 前処理： 平均特徴量を差し引く
mean_feat = np.mean(X_train_feats, axis=0, keepdims=True)
X_train_feats -= mean_feat
X_val_feats -= mean_feat
X_test_feats -= mean_feat

# 前処理： 標準偏差で割る。これにより、
# 各特徴のスケールがほぼ同じになります。
std_feat = np.std(X_train_feats, axis=0, keepdims=True)
X_train_feats /= std_feat
X_val_feats /= std_feat
X_test_feats /= std_feat

# 前処理： バイアス次元の追加
X_train_feats = np.hstack([X_train_feats, np.ones((X_train_feats.shape[0], 1))])
X_val_feats = np.hstack([X_val_feats, np.ones((X_val_feats.shape[0], 1))])
X_test_feats = np.hstack([X_test_feats, np.ones((X_test_feats.shape[0], 1))])
```

## Train SVM on features
Using the multiclass SVM code developed earlier in the assignment, train SVMs on top of the features extracted above; this should achieve better results than training SVMs directly on top of raw pixels.

この方法は、生のピクセルの上で直接SVMを訓練するよりも良い結果が得られるはずである。

```python
# 検証セットを使って学習率と正則化の強さを調整する

from cs231n.classifiers.linear_classifier import LinearSVM

results = {}
best_val = -1
best_svm = None

################################################################################
# TODO:                                                                        #
# Use the validation set to set the learning rate and regularization strength. #
# This should be identical to the validation that you did for the SVM; save    #
# the best trained classifer in best_svm. You might also want to play          #
# with different numbers of bins in the color histogram. If you are careful    #
# you should be able to get accuracy of near 0.44 on the validation set.       #
#                                                                              #
# 検証セットを使って、学習率と正則化の強さを設定する。
# これはSVMのために行った検証セットと同じでなければならない。
# また、カラーヒストグラムのビンの数を変えて遊びたくなるかもしれません。
# 注意深くやれば、検証セットで0.44に近い精度が得られるはずです。
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

learning_rates = np.geomspace(1e-4, 1e-2, 3)
regularization_strengths = np.geomspace(0.01, 1, 3)

import itertools

for lr, reg in itertools.product(learning_rates, regularization_strengths):
    # SVMを作成し、それを訓練する
    svm = LinearSVM()
    svm.train(X_train_feats, y_train, lr, reg, num_iters=1500)

    # 訓練セットと検証セットの精度を計算し、辞書に追加する。
    y_train_pred, y_val_pred = svm.predict(X_train_feats), svm.predict(X_val_feats)
    results[(lr, reg)] = np.mean(y_train == y_train_pred), np.mean(y_val == y_val_pred)

    # 検証精度が最高の場合は保存
    if results[(lr, reg)][1] > best_val:
        best_val = results[(lr, reg)][1]
        best_svm = svm

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

# 結果を表示する
for lr, reg in sorted(results):
    train_accuracy, val_accuracy = results[(lr, reg)]
    print(
        "lr %e reg %e train accuracy: %f val accuracy: %f"
        % (lr, reg, train_accuracy, val_accuracy)
    )

print("best validation accuracy achieved: %f" % best_val)
```

```python
# 学習したSVMをテストセットで評価する：少なくとも0.40が得られるはずだ。
y_test_pred = best_svm.predict(X_test_feats)
test_accuracy = np.mean(y_test == y_test_pred)
print(test_accuracy)
```

## Neural Network on image features
Earlier in this assigment we saw that training a two-layer neural network on raw pixels achieved better classification performance than linear classifiers on raw pixels. In this notebook we have seen that linear classifiers on image features outperform linear classifiers on raw pixels. 

For completeness, we should also try training a neural network on image features. This approach should outperform all previous approaches: you should easily be able to achieve over 55% classification accuracy on the test set; our best model achieves about 60% classification accuracy.

この課題の前半では、生のピクセルで2層のニューラルネットワークを学習させると、生のピクセルで線形分類器を学習させるよりも良い分類性能が得られることを見ました。このノートブックでは、画像の特徴に対する線形分類器が、生のピクセルに対する線形分類器よりも優れた性能を示すことを見てきました。

完全を期すために、画像特徴に対するニューラルネットワークの学習も試してみましょう。このアプローチはこれまでのすべてのアプローチを凌駕するはずです：テストセットで55％以上の分類精度を簡単に達成できるはずです；我々の最良のモデルは約60％の分類精度を達成しています。

```python
# Preprocessing: Remove the bias dimension
# Make sure to run this cell only ONCE
print(X_train_feats.shape)
X_train_feats = X_train_feats[:, :-1]
X_val_feats = X_val_feats[:, :-1]
X_test_feats = X_test_feats[:, :-1]

print(X_train_feats.shape)
```

```python
from cs231n.classifiers.fc_net import TwoLayerNet
from cs231n.solver import Solver

input_dim = X_train_feats.shape[1]
hidden_dim = 500
num_classes = 10

data = {
    "X_train": X_train_feats,
    "y_train": y_train,
    "X_val": X_val_feats,
    "y_val": y_val,
    "X_test": X_test_feats,
    "y_test": y_test,
}

net = TwoLayerNet(input_dim, hidden_dim, num_classes)
best_net = None

################################################################################
# TODO: Train a two-layer neural network on image features. You may want to    #
# cross-validate various parameters as in previous sections. Store your best   #
# model in the best_net variable.                                              #
#                                                                              #
# 画像の特徴で2層のニューラルネットワークを訓練する。
# 前のセクションと同様に、様々なパラメータを交差検証するとよいでしょう。
# 最良のモデルをbest_net変数に格納する。
################################################################################
# *****START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****

results = {}
best_val = -1

learning_rates = np.linspace(1e-2, 2.75e-2, 4)
regularization_strengths = np.geomspace(1e-6, 1e-4, 3)

data = {
    "X_train": X_train_feats,
    "X_val": X_val_feats,
    "X_test": X_test_feats,
    "y_train": y_train,
    "y_val": y_val,
    "y_test": y_test,
}

import itertools

for lr, reg in itertools.product(learning_rates, regularization_strengths):
    # 2層ネットを作成し、ソルバーで学習させる
    model = net
    solver = Solver(
        model, data, optim_config={"learning_rate": lr}, num_epochs=15, verbose=False
    )
    solver.train()

    # 検証セットの精度を計算し、辞書に追加する。
    results[(lr, reg)] = solver.best_val_acc

    # 検証精度が最高の場合は保存
    if results[(lr, reg)] > best_val:
        best_val = results[(lr, reg)]
        best_net = model

# 結果を表示する
for lr, reg in sorted(results):
    val_accuracy = results[(lr, reg)]
    print("lr %e reg %e val accuracy: %f" % (lr, reg, val_accuracy))

print("best validation accuracy achieved during cross-validation: %f" % best_val)

# *****END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)*****
```

```python
# 最適なニューラルネット分類器をテストセットで実行します。
# 55%以上の精度が得られるはずです。

y_test_pred = np.argmax(best_net.loss(data["X_test"]), axis=1)
test_acc = (y_test_pred == data["y_test"]).mean()
print(test_acc)
```
